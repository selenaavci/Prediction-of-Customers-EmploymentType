```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

data = pd.read_csv("/Users/selenaavci/Desktop/AutoInsurance.csv")
data.head(5)
```

```python
columns_to_drop = ['Months Since Policy Inception', 'Number of Open Complaints', 'Policy', 'Renew Offer Type', 'Coverage', 'Customer Lifetime Value', 'Location Code', 'Effective To Date', 'Months Since Last Claim','Customer']
data.drop(columns=columns_to_drop, inplace=True)
```

```python
X = data.columns.values.tolist()
X
```

```python
label_encoder = LabelEncoder()

def encoder(st):
    data[st]= label_encoder.fit_transform(data[st])
    data[st].unique()

d = data.dtypes
j = 0
for i in d:
    if(i == object):
        encoder(X[j])
    j+=1

data.head(5)
```

```python
x = data.drop(['EmploymentStatus'], axis=1)
y = data['EmploymentStatus']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000)
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
classification_report_str = classification_report(y_test, y_pred)

# Print the results
print(f'Accuracy: {accuracy:.2f}')
print('Classification Report:\n', classification_report_str)
```
