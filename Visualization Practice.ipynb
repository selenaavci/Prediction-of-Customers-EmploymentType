```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dt = pd.read_csv("/Users/selenaavci/Desktop/AutoInsurance.csv")
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(dt['Customer Lifetime Value'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Customer Lifetime Value')
plt.xlabel('Customer Lifetime Value')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.countplot(x='Marital Status', hue='EmploymentStatus', data=dt, palette='Set2')
plt.title('Count of Employment Status by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='EmploymentStatus', y='Income', data=dt, palette='Set1')
plt.title('Boxplot of Income by Employment Status')
plt.xlabel('Employment Status')
plt.ylabel('Income')
plt.show()
```

```python
numerical_cols = dt.select_dtypes(include=['int64', 'float64']).columns
dt[numerical_cols].hist(bins=20, figsize=(10, 10))
plt.suptitle('Histogram Grid for Numerical Features', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='State', hue='Response', data=dt, palette='Set1')
plt.title('Count of Responses from States')
plt.xlabel('State')
plt.ylabel('Count')
plt.legend(title='Response', loc='upper right')
plt.show()
```

```python
sns.pairplot(dt, hue='EmploymentStatus', palette='viridis')
plt.suptitle('Pairplot by Employment Status', y=1.02)
plt.show()
```
